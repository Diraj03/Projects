```python
d=d.drop(["ID","Var_1"],axis=1)
e=e.drop(["ID","Var_1"],axis=1)
```

```python
sb.distplot(d["Age"])
```

```python
d.fillna({'Work_Experience':0,
          'Family_Size':0},inplace=True)

e.fillna({'Work_Experience':0,
          'Family_Size':0},inplace=True)
```

```python
d["Graduated"].fillna(d["Graduated"].mode()[0],inplace=True)
d["Ever_Married"].fillna(d["Ever_Married"].mode()[0],inplace=True)
d["Profession"].fillna(d["Profession"].mode()[0],inplace=True)
    
```

```python
e["Graduated"].fillna(e["Graduated"].mode()[0],inplace=True)
e["Ever_Married"].fillna(e["Ever_Married"].mode()[0],inplace=True)
e["Profession"].fillna(e["Profession"].mode()[0],inplace=True)
```

```python
a=d.iloc[:,:-1]
b=d.iloc[:,-1]
```

```python
a1=e.iloc[:,:-1]
b1=e.iloc[:,-1]
```

```python
x=pd.get_dummies(a,drop_first=True)
```

```python
y=pd.get_dummies(a1,drop_first=True)
```

```python
x
```

```python
y
```

```python
train=x.iloc[::].values
test=y.iloc[::].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(train)
x_test=sc.transform(test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=0)
```

```python
classifier.fit(x_train,b)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(b1,y_pred)
```

```python
accuracy_score(b1,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={"n_estimators":[10,20,50,75,100],
            "criterion":["gini","entropy"],
            "bootstrap":[True,False]}
```

```python
classifier=RandomForestClassifier(random_state=0)
```

```python
gr=GridSearchCV(estimator=classifier,param_grid=param_grid,cv=5,n_jobs=-1)
```

```python
gr.fit(train,b)
```
